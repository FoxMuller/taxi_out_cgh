```python
import os
os.environ["OPENSKY_USERNAME"] = "username"
os.environ["OPENSKY_PASSWORD"] = "password"

from pyopensky.trino import Trino

trino = Trino()
df = trino.flightlist(
        start=f"2025-11-01",
        stop=f"2025-12-01",   # stop é exclusivo, cobre 01 e 02 inteiros
        airport="SBSP",
    )
```

```python
df.to_parquet(f"data/flightlist/fl_202511.parquet")
```
